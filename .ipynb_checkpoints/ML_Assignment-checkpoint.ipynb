### Imputing Nuemrical Data

```python
train_X_no_nans =  train_X.copy()

# handling missing numerical data

# Imputer
imptr = SimpleImputer(missing_values = np.nan, strategy = 'mean')  


# Fit the imputer object to the feature matrix - only the numeric features
imptr = imptr.fit(train_X_no_nans[['ApplicantIncome', 'CoapplicantIncome',
                'LoanAmount', 'Loan_Amount_Term', 'Credit_History']])

# Call Transform to replace missing data in dataset by the mean of the column to which that missing data belongs to
train_X_no_nans[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']] = \
imptr.transform(train_X_no_nans[['ApplicantIncome', 'CoapplicantIncome','LoanAmount', 'Loan_Amount_Term', 'Credit_History']]) 

# to check that indeed the NaNs have been filled with values
print(train_X_no_nans[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']].isnull().sum())

train_X_no_nans
```

```python
#repeating the same process for the test split

test_X_no_nans =  test_X.copy()

# handling missing numerical data

# Imputer
imptr = SimpleImputer(missing_values = np.nan, strategy = 'mean')  


# Fit the imputer object to the feature matrix - only the numeric features
imptr = imptr.fit(test_X_no_nans[['ApplicantIncome', 'CoapplicantIncome',
                'LoanAmount', 'Loan_Amount_Term', 'Credit_History']])

# Call Transform to replace missing data in dataset by the mean of the column to which that missing data belongs to
test_X_no_nans[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']] = \
imptr.transform(test_X_no_nans[['ApplicantIncome', 'CoapplicantIncome','LoanAmount', 'Loan_Amount_Term', 'Credit_History']]) 

# to check that indeed the NaNs have been filled with values
print(test_X_no_nans[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']].isnull().sum())

test_X_no_nans
```

```python
#repeating the same process for the loaded test data provided initially

final_test_data_no_nans =  final_test_data.copy()

# handling missing numerical data

# Imputer
imptr = SimpleImputer(missing_values = np.nan, strategy = 'mean')  


# Fit the imputer object to the feature matrix - only the numeric features
imptr = imptr.fit(final_test_data_no_nans[['ApplicantIncome', 'CoapplicantIncome',
                'LoanAmount', 'Loan_Amount_Term', 'Credit_History']])

# Call Transform to replace missing data in dataset by the mean of the column to which that missing data belongs to
final_test_data_no_nans[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']] = \
imptr.transform(final_test_data_no_nans[['ApplicantIncome', 'CoapplicantIncome','LoanAmount', 'Loan_Amount_Term', 'Credit_History']]) 

# to check that indeed the NaNs have been filled with values
print(final_test_data_no_nans[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']].isnull().sum())

final_test_data_no_nans
```

### Imputing Categorical and Ordinal Data

```python
# handling missing categorical and ordinal data

#Imputer
imptr = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')  

# Fit the imputer object to the feature matrix - only the categorical features
imptr = imptr.fit(train_X_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']])

# Call Transform to replace missing data in dataset by the most frequent of the column to which that missing data belongs to
train_X_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']] = imptr.transform(train_X_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']]) 


# to check that indeed the NaNs have been filled with values
print(train_X_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']].isnull().sum())

train_X_no_nans
```

```python
#repeating the same process for the test split

#Imputer
imptr = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')  

# Fit the imputer object to the feature matrix - only the categorical features
imptr = imptr.fit(test_X_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']])

# Call Transform to replace missing data in dataset by the most frequent of the column to which that missing data belongs to
test_X_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']] = imptr.transform(test_X_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']]) 


# to check that indeed the NaNs have been filled with values
print(test_X_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']].isnull().sum())

test_X_no_nans
```

```python
#repeating the same process for the loaded test data provided initially

#Imputer
imptr = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')  

# Fit the imputer object to the feature matrix - only the categorical features
imptr = imptr.fit(final_test_data_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']])

# Call Transform to replace missing data in dataset by the most frequent of the column to which that missing data belongs to
final_test_data_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']] = imptr.transform(final_test_data_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']]) 


# to check that indeed the NaNs have been filled with values
print(final_test_data_no_nans[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']].isnull().sum())

test_X_no_nans
```

## Ordered Label Encoding of the Oridinal Variables

#### Note

All the ordinal variables (Dependents, Education and Property_Area) are converted to ordered label encoding to develop the ML models.

For "Dependents", we can say that 0 is at the lowest point of the scale while 3+ is at the highest point of the scale.

For "Education", we can say that "Not Graduate" is at the lowest point of the scale while "Graduate" is at the highest point of the scale.

For "Property_Area", we can say that the "Rural" is at the lowest point of the scale while "Urban" is at the highest point of the scale.

```python
#Ordered label encoding of the Oridinal variables in training split

dependent_mapper = {"0":0, "1":1, "2":2, "3+":3}
train_X_no_nans["Dependents"] = train_X_no_nans["Dependents"].replace(dependent_mapper)

education_mapper = {"Not Graduate":0, "Graduate":1}
train_X_no_nans["Education"] = train_X_no_nans["Education"].replace(education_mapper)

property_area_mapper = {"Rural":0, "Semiurban":1, "Urban":2}
train_X_no_nans["Property_Area"] = train_X_no_nans["Property_Area"].replace(property_area_mapper)

train_X_no_nans
```

```python
#Min-Max Normalization of all the converted Numerical variables for train split

train_X_no_nans[['Dependents', 'Education','Property_Area', 'Credit_History']]=min_max.fit_transform(train_X_no_nans[['Dependents', 'Education','Property_Area', 'Credit_History']])

train_X_no_nans
```

```python
#Ordered label encoding of the Oridinal variables in testing split

dependent_mapper = {"0":0, "1":1, "2":2, "3+":3}
test_X_no_nans["Dependents"] = test_X_no_nans["Dependents"].replace(dependent_mapper)

education_mapper = {"Not Graduate":0, "Graduate":1}
test_X_no_nans["Education"] = test_X_no_nans["Education"].replace(education_mapper)

property_area_mapper = {"Rural":0, "Semiurban":1, "Urban":2}
test_X_no_nans["Property_Area"] = test_X_no_nans["Property_Area"].replace(property_area_mapper)

test_X_no_nans
```

```python
#Min-Max Normalization of all the converted Numerical variables for test split

test_X_no_nans[['Dependents', 'Education','Property_Area', 'Credit_History']]=min_max.fit_transform(test_X_no_nans[['Dependents', 'Education','Property_Area', 'Credit_History']])

test_X_no_nans
```

```python
#Ordered label encoding of the Oridinal variables in the loaded test data provided initially

dependent_mapper = {"0":0, "1":1, "2":2, "3+":3}
final_test_data_no_nans["Dependents"] = final_test_data_no_nans["Dependents"].replace(dependent_mapper)

education_mapper = {"Not Graduate":0, "Graduate":1}
final_test_data_no_nans["Education"] = final_test_data_no_nans["Education"].replace(education_mapper)

property_area_mapper = {"Rural":0, "Semiurban":1, "Urban":2}
final_test_data_no_nans["Property_Area"] = final_test_data_no_nans["Property_Area"].replace(property_area_mapper)

final_test_data_no_nans
```

```python
#Min-Max Normalization of all the converted Numerical variables for the loaded test data provided initially

final_test_data_no_nans[['Dependents', 'Education','Property_Area', 'Credit_History']]=min_max.fit_transform(final_test_data_no_nans[['Dependents', 'Education','Property_Area', 'Credit_History']])

final_test_data_no_nans
```

# ML Model Development and Evaluation

#### Note

We decided to use Support Vector Classifier (SVC) and the Random Forest Classifier as this is a classifiation problem where we have to predict the 2 classes of the "Loan Status" label.

Since SVC tries to find the best hyper plane between the classes available in the data set using support vectors (even using the kernal trick in higher dimension as we have more than 2 features) we selected this as one of the ML algorithms to model the solution.

Since Random Forest Classfier uses an ensemble of decision tress based on the features available in the data set and we can get a weighted decision of different decision tree outputs using the Random Forest, we decided to use this as our second ML algorithm to test.

## SVC ML Model Development and Evaluation

```python
#SVC model training 

model  = svm.SVC(class_weight='balanced', random_state=SEED)
#here class_weight is given as "balanced" to automatically calulate the weights according to the available examples for each class
model.fit(train_X_no_nans,train_y_encoded)

ytr_p = model.predict(train_X_no_nans)
yts_p = model.predict(test_X_no_nans)
```

```python

print("Confusion Matric for the Training Set Evaluation of SVC")
cm = confusion_matrix(train_y_encoded, ytr_p)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['N', 'Y'])
disp.plot()
plt.show()
```

# Optimization of best performing ML model by Hyper Parameter Tuning

#### Note

For hyper parameter tuning, Random Forest Classifier was chosen as it gave the best results compared to the SVC.

For hyper parameter tuning of the Random Forest Classifier, the hyper parameters of 'n_estimators' which gives the number of decision trees in the Random Forest, 'max_features' which gives the no. of features to be considered when looking for the best split in a tree and 'criterion' which gives the function which measures the quality of a split in a tree were chosen.

Sklearn Grid Search with Cross Validation was used to find the best tuned parameter values for the above mentioned hyper parameters of the Random Forest Classifier using 5-fold csross validation.

Since the data is calss imbalanced as the scoring function, the weighted per label F1 score was chosen to better represent the performance compared to both the classes to avoid the class imbalance issue.

```python
rf_clf_1=RandomForestClassifier(class_weight="balanced",  random_state=SEED )
```

```python
param_grid_rfc = { 
    'n_estimators': [10, 50],
    'max_features': ['auto', 'sqrt', 'log2'],
    'criterion' :['gini', 'entropy', 'log_loss']
}
```

```python
grid_rfc = GridSearchCV(estimator=rf_clf_1, param_grid=param_grid_rfc, scoring='f1_weighted', cv=5)
grid_rfc.fit(train_X_no_nans,train_y_encoded)
```

```python
grid_rfc.best_params_
```

```python
rf_clf_2=RandomForestClassifier(n_estimators=50, criterion='gini', max_features='auto', class_weight="balanced", random_state=SEED )

rf_clf_2.fit(train_X_no_nans,train_y_encoded)

ytr_p = rf_clf_2.predict(train_X_no_nans)
yts_p = rf_clf_2.predict(test_X_no_nans)
```

```python
print('\nRandom Forest New Training Metrics\n')
print(classification_report(train_y_encoded, ytr_p, target_names=['N', 'Y'], digits=4))
```

```python
print('\nRandom Forest New Testing Metrics\n')
print(classification_report(test_y_encoded, yts_p, target_names=['N', 'Y'], digits=4))
```

#### Note

using the hyper parameters found in the grid search wehn we train the Random Forest Classifier again there is an improvement of 1.57% in accuracy and an increase of 1.76% in weighted F1 Score.

But still the data seem to be overfiting as the training accuracy is 100%.
